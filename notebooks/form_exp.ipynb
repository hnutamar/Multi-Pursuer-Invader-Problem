First experiment is done with 5 pursuers. The maximum acceleration of pursuers will be 1.0, acceleration of Prime unit is only 0.1, which in current settings makes pursuers 3 times faster than Prime. That is why pursuers should not have a problem with keeping the formation intact.

```python
PURSUER_NUM = 5
INVADER_NUM = 0
PURS_ACC = 1.0
PRIME_ACC = 0.3

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC)
sim.run()
```

When we increase number of pursuers to for example 20, pursuers start to have problem with avoid all the other pursuers before settling to the formation. This is caused slightly by the random positional initialization. That can lead to pursuers flying too fast to formation, but because of the strong repulsive force inside of the formation, they are immediately being thrown outside, where other pursuers are. Once pursuers settle in their position, probability of collision is much lower.

```python
PURSUER_NUM = 20
INVADER_NUM = 0
PURS_ACC = 1.0
PRIME_ACC = 0.3

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC)
sim.run()
```

Now, what if Prime unit flies faster? This can definitely pose a problem for pursuers, whose algorithm, by now, rely only on vortex field, which forces them to circle around the Prime.
