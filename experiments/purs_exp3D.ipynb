### Pursuing with obstacle

In this experiment, obstacle is in the way of Prime. Pursuers therefore need to eliminate invaders and also push Prime out of a way from obstacle. In 3D, this job is much harder, because Prime has more possibilities where to go, that means more possibilities to go the wrong way. Pursuers compute which direction is the best to push Prime so that it would not crash into obstacle. This direction is then fixed a pursuers are not going to change it.
