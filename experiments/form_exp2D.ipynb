### Basic formation

First experiment is done with 5 pursuers. The maximum acceleration of pursuers will be 2.0, acceleration of Prime unit is only 0.1, which in current settings makes pursuers much faster than Prime. That is why pursuers should not have a problem with keeping the formation intact.

```python
PURSUER_NUM = 5
INVADER_NUM = 0
PURS_ACC = 2.0
PRIME_ACC = 0.1
prime_pos = np.array([3.0, 3.0])
purs_pos = np.array([[5.0, 5.0], [0.0, 0.0], [5.0, 0.0], [0.0, 5.0], [3.0, 5.0], [0.0, 3.0]])
#crash enabled from the beginning
CRASH_ENABLED = 1

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM, world_height=30, world_width=30)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC, 
                    prime_pos=prime_pos, purs_pos=purs_pos, crash_enabled=CRASH_ENABLED, formation_delay=1)
sim.run()
```

### Obstacle avoidance

Next scenario introduces new feature - obstacle. In this settings, Prime is flying directly into the obstacle, which Prime does not see. Pursuers therefore must adjust Prime's path, so that Prime would not crash into the obstacle. Correct behavior of this relies on well tuned parameters, so it is very unpredictable. However, most of a time the adjustment of path works well.

```python
from prime_mode import Modes

PURSUER_NUM = 10
INVADER_NUM = 0
PURS_ACC = 3.0
PRIME_ACC = 0.2
OBS_RAD = 2.0
prime_pos = np.array([3.0, 3.0])
purs_pos = np.array([[7.0, 7.0], [-2.0, -2.0], [7.0, -2.0], [-2.0, 7.0], [7.0, 4.0], [7.0, 1.0], [-2.0, 1.0], [-2.0, 4.0], 
                     [1.0, 7.0], [4.0, 7.0]])
#crash enabled from the beginning
CRASH_ENABLED = 1

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM, world_height=30, world_width=30, obstacle=True, obstacle_rad=2.0)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC, prime_pos=prime_pos, 
                      purs_pos=purs_pos, crash_enabled=CRASH_ENABLED, formation_delay=10, prime_mode=Modes.LINE)
sim.run()
```

You can see, that pursuers change their direction in formation according to which way it is easier to push Prime.
