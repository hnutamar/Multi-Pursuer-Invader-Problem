### Basic formation

Because 3D world allows more space, experiments can be done with more pursuers. In this experiment, 30 pursuers forms around the prime. Consensus on the direction is also used, because drones have different internal clock. That is done only by observing surrounding drones and computing weighted sum of desired direction and direction from the average directions of other drones. This approach should work even if internal clocks of pursuers have different speeds.

### Obstacle avoidance
