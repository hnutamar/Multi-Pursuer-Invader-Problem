First experiment is done with 5 pursuers. The maximum acceleration of pursuers will be 2.0, acceleration of Prime unit is only 0.1, which in current settings makes pursuers 6 times faster than Prime. That is why pursuers should not have a problem with keeping the formation intact.

```python
PURSUER_NUM = 6
INVADER_NUM = 0
PURS_ACC = 2.0
PRIME_ACC = 0.1
prime_pos = np.array([3.0, 3.0])
purs_pos = np.array([[5.0, 5.0], [0.0, 0.0], [5.0, 0.0], [0.0, 5.0], [3.0, 5.0], [0.0, 3.0]])
#crash enabled from the beginning
CRASH_ENABLED = 1

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM, world_height=30, world_width=30)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC, 
                    prime_pos=prime_pos, purs_pos=purs_pos, crash_enabled=CRASH_ENABLED, formation_delay=1)
sim.run()
```

### Prime flying in circles

How the formation behaves can be also nicely seen on this example. Instead of prime flying in straight line, it flies in circles. This means that the formation changes dynamically according to the prime velocity vector.

```python
from prime_mode import Modes

PURSUER_NUM = 20
INVADER_NUM = 0
PURS_ACC = 1.0
PRIME_ACC = 0.1
prime_pos = np.array([3.0, 3.0])
purs_pos = np.array([[7.0, 7.0], [-2.0, -2.0], [7.0, -2.0], [-2.0, 7.0], [7.0, 4.0], [7.0, 1.0], [-2.0, 1.0], [-2.0, 4.0], 
                     [1.0, 7.0], [4.0, 7.0], [1.0, -2.0], [4.0, -2.0], [5.0, 5.0], [0.0, 0.0], [5.0, 0.0], [2.0, 5.0], 
                     [3.0, 5.0], [0.0, 3.0], [9.0, 9.0], [9.0, 0.0]])
#crash enabled from the beginning
CRASH_ENABLED = 1

conf = Sim2DConfig(purs_num=PURSUER_NUM, inv_num=INVADER_NUM, world_height=30, world_width=30)
sim = DroneSimulation(conf, purs_acc=PURS_ACC, prime_acc=PRIME_ACC, prime_pos=prime_pos, 
                      purs_pos=purs_pos, crash_enabled=CRASH_ENABLED, formation_delay=1, prime_mode=Modes.CIRCLE)
sim.run()
```
